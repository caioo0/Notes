## 矩阵乘法

知识点：

- 矩阵乘向量可以看做是对A的列的线性组合
- 矩阵乘以右侧列向量可看成矩阵各列向量的线性组合，结果为列向量！
- 左侧行向量乘以矩阵可看成矩阵各行向量的线性组合，结果为行向量！


两个矩阵相乘得到第三个矩阵

$$
 \bf C=AB
$$
					
为了使得定义合法，我们需要 $\mathbf A$ 的形状为 $m\times n$, $\mathbf B$ 的形状为 $n\times p$，得到的矩阵为 $\mathbf C$ 的形状为 ${m\times p}$。

定义为

$$
C_{i,j} = \sum_{k} A_{i,k}B_{k,j}
$$
		
注意矩阵乘法不是逐元素相乘，逐元素相乘又叫 `Hadamard` 乘积，记作 $\bf A\odot B$。

向量可以看出是列为 $1$ 的矩阵，两个相同大小的向量 $\bf x, y$ 的点乘（`dot product`）或者内积，可以使用矩阵乘法表示为 $\bf x^\top y$。

我们也可以把矩阵乘法理解为： $C_{i,j}$ 表示 $\bf A$ 的第 $i$ 行与 $\bf B$ 的第 $j$ 列的点乘。

#### 列乘以行

在常规方法中，计算$A*B=C$矩阵乘法时，使用$A$的行向量乘上$B$的列向量得到$C$中各个位置的元素。

我们这次介绍的方法，是用A的列向量乘上B的行向量得到各个矩阵，再将矩阵相加，得到C。

【例】求解 $\begin{bmatrix}2 & 7 \\ 3 & 8 \\ 4 & 9 \end{bmatrix}\begin{bmatrix}1 & 6 \\ 0 & 0 \end{bmatrix}$

列乘以行方法：

$$
\begin{bmatrix}2 & 7 \\ 3 & 8 \\ 4 & 9 \end{bmatrix}\begin{bmatrix}1 & 6 \\ 0 & 0 \end{bmatrix}
=
\begin{bmatrix}2 \\ 3 \\ 4 \end{bmatrix}\begin{bmatrix}1 & 6 \end{bmatrix}
+ 
\begin{bmatrix}7 \\ 8 \\ 9 \end{bmatrix}\begin{bmatrix}0 & 0 \end{bmatrix}
=
\begin{bmatrix}2 & 12\\ 3 & 18 \\ 4 & 24 \end{bmatrix}
$$

#### 分块做乘法

分块乘法就是宏观上的矩阵乘法，比如现在有一个$50*50$的矩阵与$50*50$矩阵相乘，一个一个进行运算很麻烦，尤其是如果矩阵在某一个区域上有一定的性质，那么我们可以将其分块，如：

$$
\begin{bmatrix}A1 & A2 \\ A3 & A4 \end{bmatrix}\begin{bmatrix}B1 & B2 \\ B3 & B4\end{bmatrix}
=
\begin{bmatrix}C1 & C2\\ C3 & C4 \end{bmatrix}
$$

其中的$A_{1,2,3,4}$与$B_{1,2,3,4}$都是划分之后的一块块矩阵，如：

$$
\left[
\begin{array}{c|c}
?& ? \\ \hline
?& ?
\end{array}
\right]
$$
得到的各个部分。而$C_1=A_1B_1 + A_2B_3$,和矩阵乘法的计算步骤一样，只是这里的$A_1B_1,A_2B_3$都是矩阵之间的乘法而已。只要$A$ 和 $B$ 分块相互匹配，就可以用这样的分块乘法求解。

通用公式：

$$
\left[\begin{array}{c|c}A_1&A_2\\\hline A_3&A_4\end{array}\right]\left[\begin{array}{c|c}B_1&B_2\\\hline B_3&B_4\end{array}\right]=\left[\begin{array}{c|c}A_1B_1+A_2B_3&A_1B_2+A_2B_4\\\hline A_3B_1+A_4B_3&A_3B_2+A_4B_4\end{array}\right]
=
\left[\begin{array}{c|c}C1 & C2\\ \hline C3 & C4 \end{array}\right]
$$



## 逆矩阵

### 逆矩阵介绍

之前介绍过一点逆矩阵的定义，对于一个方阵$A$, 如果$A$可逆，就有这样一个$A^{-1}$使
$$
AA^{-1} = I = A^{-1}A
$$

如果A是**非方阵**，左侧的$A^{-1}$与右侧的$A^{-1}$不会相同，因为这时左右侧$A^{-1}$形状不相同，那么违背了我们说的"一个"$A^{-1}$的情况。


> **若存在非零向量$x$,使得$Ax=0$,那么$A$就不可能有逆矩阵**

证明：如果A有逆，在Ax=0这个等式两端同时乘上$A^{-1}$,有：

$$
A^{-1}Ax = Ix = 零向量
$$
而$Ix$不可能是零向量，这种情况下，A没有逆矩阵成立。

证明矩阵是否存在逆矩阵：
$$\begin{bmatrix}1 & 3 \\ 2 & 6\end{bmatrix}$$

- 方法1：这个矩阵的行列式为0，所以矩阵不可逆。 

- 方法2：矩阵包含两个列向量：
$\begin{bmatrix}1 \\ 2 \end{bmatrix},\begin{bmatrix}3 \\ 6\end{bmatrix}$ 他们之间互成倍数，也就是两个向量之一对其线性组合无意义，那么这个$A$不可逆。


- 方法3：$$\begin{bmatrix}1 & 3 \\ 2 & 6\end{bmatrix}\begin{bmatrix}3\\-1\end{bmatrix} =\begin{bmatrix}0\\0\end{bmatrix} $$ 

证明：如果对于非零的$x$仍有$Ax=0$，而$A$有逆$A^{−1}$，则$A^{−1}Ax=0$，即$x=0$，与题设矛盾，得证，矩阵不可逆；


### 逆矩阵求解

其实求逆矩阵就是解方程组的过程，举例说明：

【例】设$A=\begin{bmatrix}1 & 3 \\ 2 & 7\end{bmatrix}$,求逆矩阵$A^{-1}$

$$
\begin{bmatrix}1 & 3 \\ 2 & 7\end{bmatrix}\begin{bmatrix}a & b \\ c & d\end{bmatrix} = I = \begin{bmatrix}1 & 0 \\ 0 & 1\end{bmatrix} 
$$

高斯-若尔当（Gauss-Jordan）方法，该方法可以一次处理所有的方程：

具体操作方法即使用增广矩阵将虚线左侧消为单位矩阵$I$,此时右侧矩阵即为逆矩阵$A^{-1}$；

$$
\left[\begin{array}{cc:cc}1 & 3 & 1 & 0 \\ 2 & 7 & 0 & 1\end{array}\right] 
\xrightarrow{row_2-2row_1}
\left[\begin{array}{cc:cc}1 & 3 & 1 & 0 \\ 0 & 1 & -2 & 1\end{array}\right]
\xrightarrow{row_1-3row_2} 
\left[\begin{array}{cc:cc}1 & 0 & 7 & -3 \\ 0 & 1 & -2 & 1\end{array}\right]
$$

我们就将矩阵从$[A | I]$变为$[I | A^{−1}]$

$$
A^{-1} = \left[\begin{array}{cc} 7 & -3 \\  -2 & 1\end{array}\right]
$$
通过$AA^{-1} = I $检验一下，正确。

高斯-若尔当法的本质是使用消元矩阵$E$，对A进行操作，$E\left[\begin{array}{c|c}A&I\end{array}\right]$，利用一步步消元有$EA=I$，进而得到$\left[\begin{array}{c|c}I&E\end{array}\right]$，其实这个消元矩阵$E$就是$A−1$，而高斯-若尔当法中的$I$只是负责记录消元的每一步操作，待消元完成，逆矩阵就自然出现了。
