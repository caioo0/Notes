2.用plot方法画出x=(0,10)间sin的图像

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,300)

plt.plot(x, np.sin(x)) 
```

3.用点加线的方式画出x=(0,10)间sin的图像

```python
x = np.linspace(0,10,30)
plt.plot(x, np.sin(x), linestyle='--',marker='o') 
plt.show()
```

4.用scatter方法画出x=(0,10)间sin的点图像


```python
import matplotlib as mpl

x = np.linspace(0,10,30)
#plt.scatter(x, np.sin(x), marker='o') 
y = np.random.randn(30)
#设置画布大小
plt.figure(figsize = (10, 6))

#设置颜色
colors = np.random.randn(30)

plt.scatter(x, y, 
            #大小
            s = np.power(3*x + 4*y, 2),
            #颜色
            c = colors,
            #标注
            marker = 'o',
            #指定某个colormap值
            cmap=mpl.cm.RdYlBu,
            #透明度
            alpha = 0.9)
plt.show()

```

5.用饼图的面积及颜色展示一组4维数据

```python
rng = np.random.RandomState(0)

x = rng.randn(100)
y = rng.randn(100)

colors = rng.rand(100)

# !@ numpy.random.randn(d0, d1, …, dn)是从标准正态分布中返回一个或多个样本值。
# !@ numpy.random.rand(d0, d1, …, dn)的随机样本位于[0, 1)中。

sizes = 1000* rng.rand(100)

plt.scatter(x,y,c=colors,s=sizes,alpha=0.3,cmap='viridis')
plt.colorbar()
```

### 附加


数据生成

生成1024个呈标准正态分布的二维数据组 (平均数是0，方差为1) 作为一个数据集，并图像化这个数据集。每一个点的颜色值用T来表示：

    正态分布是一种概率分布。正态分布是具有两个参数μ和σ的连续型随机变量的分布，第一参数μ是服从正态分布的随机变量的均值，第二个参数σ是此随机变量的方差，所以正态分布记作N(μ，σ )。
    
正态分布写法： `np.random.normal(loc=0.0, scale=1.0, size=None)` ,其中`loc`为均值，`scale`为标准差

```python
import matplotlib.pyplot as plt 
import numpy as np 
n = 1024

x = np.random.normal(0,1,n) 
y = np.random.normal(0,1,n)
plt.figure(figsize=(20, 10), dpi=100) # 创建画布
plt.hist(x,50)
plt.hist(y,50)
plt.show()
```

```python
T = np.arctan2(y,x) # for color value
```

```python
plt.scatter(x, y, s=75, c=T, alpha=.5)

plt.xlim(-1.5, 1.5)
plt.xticks(())  # ignore xticks
plt.ylim(-1.5, 1.5)
plt.yticks(())  # ignore yticks

plt.show()
```
